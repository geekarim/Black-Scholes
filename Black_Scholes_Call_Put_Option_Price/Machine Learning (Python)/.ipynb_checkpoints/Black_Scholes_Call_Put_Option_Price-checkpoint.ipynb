```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
#import torch.optim as optim
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from scipy.stats import norm
```

```python
torch.cuda.empty_cache()
```

```python
class Black_Scholes_Call_Put():
    '''
    Class for handling Black-Scholes option pricing with neural networks.

    Attributes:
        n_epochs (int): Number of training epochs.
        n_data (int): Number of data points.
        device (torch.device): Device to run computations on (CPU or GPU).
        network (Black_Scholes_Call_Put_Network): Neural network model for option pricing.
        MSE_loss (torch.nn.MSELoss): Mean squared error loss function for training.
    '''
    
    def __init__(self):
        '''
        Initializes Black_Scholes_Call_Put with default parameters and sets up necessary components.
        '''
        super(Black_Scholes_Call_Put,self).__init__()
       
        self.n_epochs = 5  # Default number of epochs for training
        self.n_data = 10000  # Default number of data points

        # Determine the device for computation (CUDA if available, otherwise CPU)
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        # Set random seeds for reproducibility on GPU
        torch.manual_seed(0)
        torch.cuda.manual_seed(0)
        torch.cuda.manual_seed_all(0)
        torch.backends.cudnn.benchmark = False
        torch.backends.cudnn.deterministic = True

        # Initialize the neural network model
        self.network = Black_Scholes_Call_Put_Network()

        # Define the loss function (Mean Squared Error)
        self.MSE_loss = nn.MSELoss(reduction='sum')

        # There are no parameters to optimize here initially
        # self.optimizer = torch.optim.Adam(self.network.parameters(), lr=1e-2)
        
    def load_model(self, savepath):
        '''
        Loads a pre-trained model from the specified file path.

        Args:
            savepath (str): The path to the file where the pre-trained model is saved.
        
        Note:
            This method should load the model's state dictionary or weights from the file specified by `savepath`.
            The actual loading mechanism (e.g., `torch.load`) should be implemented here.
        '''
        pass

    def save_model(self, savepath):
        '''
        Saves the trained model's weights to the specified file path.

        Args:
            savepath (str): The path to the file where the model's weights will be saved.
        
        Note:
            This method should save the model's state dictionary or weights to the file specified by `savepath`.
            The actual saving mechanism (e.g., `torch.save`) should be implemented here.
        '''
        pass

    
    def Black_Scholes_Call_Put_Option_Price(self,S,K,r,T,sigma):
        '''Computes Black-Scholes Call & Put Option Prices

        Input:
              S (float matrix) = Stock Price
              K (float matrix) = Strike
              r (float) = Interest Rate
              T (float) = Time to Maturity
              sigma (float) = Volatility

        Output:
              Call = Black-Scholes Call Option Price
              Put = Black-Scholes Put Option Price
        '''

        if T == 0:
            Call = max( S - K , 0 )
            Put = max( K - S , 0 )
        else:
            d1 = (np.log(S/K) + (r+sigma**2 / 2)*T)/(sigma*np.sqrt(T))
            d2 = (np.log(S/K) + (r-sigma**2 / 2)*T)/(sigma*np.sqrt(T))
            Call = np.multiply(S,norm.cdf(d1)) - np.exp(-r*T)*np.multiply(K,norm.cdf(d2))
            Put = np.exp(-r*T)*np.multiply(K,norm.cdf(-d2)) - np.multiply(S,norm.cdf(-d1))

        return Call, Put
    
    def generate_data(self):
        '''
        Generates synthetic data for training and testing the Black-Scholes option pricing model.

        The data consists of the following features:
        - X_S: Stock price (random integers between 1 and 100)
        - X_K: Strike price (random integers between 1 and 100)
        - X_r: Risk-free interest rate (random floats between 0 and 0.1)
        - X_T: Time to maturity (random integers between 0 and 10)
        - X_sigma: Volatility (random floats between 0.01 and 0.2)

        The target variable Y is the Call and Put option prices calculated using the Black-Scholes formula.

        After generating the dataset, it splits it into training and test sets (90% training, 10% testing).
        The data is then converted to PyTorch tensors for model training.

        Returns:
            None
        '''
        
        # Generate random features: stock price (X_S), strike price (X_K), interest rate (X_r), 
        # time to maturity (X_T), and volatility (X_sigma)
        X_S = np.random.randint(1, 101, size=self.n_data)  # Stock prices between 1 and 100
        X_K = np.random.randint(1, 101, size=self.n_data)  # Strike prices between 1 and 100
        X_r = np.random.randint(0, 21, size=self.n_data) / 100  # Interest rates between 0 and 0.2
        X_T = np.random.randint(0, 11, size=self.n_data)  # Time to maturity between 0 and 10 years
        X_sigma = np.random.randint(1, 51, size=self.n_data) / 100  # Volatility between 0.01 and 0.5

        # Stack the features into a single array (shape: [n_data, 5])
        X = np.stack((X_S, X_K, X_r, X_T, X_sigma), axis=-1)

        # Calculate the corresponding options prices using the Black-Scholes formula
        Y = []
        for i in range(self.n_data):
            # Calculate options price for each set of inputs
            Y.append(self.Black_Scholes_Call_Put_Option_Price(*list(X[i])))

        # Split the data into training and testing sets (90% train, 10% test)
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=0)

        # Convert the data to PyTorch tensors for model training
        self.X_train = torch.FloatTensor(X_train)
        self.Y_train = torch.FloatTensor(Y_train)
        self.X_test = torch.FloatTensor(X_test)
        self.Y_test = torch.FloatTensor(Y_test)
        
    def train(self):
        '''
        Trains the Black-Scholes neural network model using the generated data.

        The training loop iterates over a predefined number of epochs (self.n_epochs). In each epoch, it:
        1. Generates fresh data for training and testing.
        2. Makes predictions using the network on the training data.
        3. Calculates the Mean Squared Error (MSE) loss between the predictions and the true values.
        4. Optionally, updates the model parameters (commented out in this version).
        5. Evaluates the model on the test data and records the loss for each epoch.
        6. Appends the evaluation loss to self.Train_loss and prints the training progress.

        After training is completed, the method invokes self.plot_training() to visualize the training loss.
        '''
        
        self.Train_loss = []  # List to store the training loss for each epoch
        
        def tuple_of_tensors_to_tensor(tuple_of_tensors):
            '''
            Converts a tuple of tensors into a single stacked tensor.

            Args:
                tuple_of_tensors (tuple): A tuple of tensors.
            
            Returns:
                torch.Tensor: A single stacked tensor created from the tuple.
            '''
            return torch.stack(list(tuple_of_tensors), dim=0)
        
        # Training loop for the specified number of epochs
        for epoch in tqdm(range(self.n_epochs)):
            # Generate new data for the current epoch
            self.generate_data()
            
            # Zero the gradients (if an optimizer is used, this line would reset the gradients)
            # self.optimizer.zero_grad()

            # Make predictions using the network on the training data
            predictions = tuple_of_tensors_to_tensor(self.network(self.X_train))

            # Compute the Mean Squared Error (MSE) loss on the training data
            loss = self.MSE_loss(predictions.to(self.device), self.Y_train.to(self.device))
            loss = loss / predictions.shape[0]  # Normalize the loss by the number of predictions
            
            # Backpropagate the loss and update the model parameters (commented out here)
            # loss.backward()
            # self.optimizer.step()

            # Evaluation phase: evaluate the model on the test set (without backpropagation)
            with torch.no_grad():
                eval_predictions = tuple_of_tensors_to_tensor(self.network(self.X_test))

                # Compute the loss on the test data
                eval_loss = self.MSE_loss(eval_predictions.to(self.device), self.Y_test.to(self.device))
                eval_loss = eval_loss / eval_predictions.shape[0]  # Normalize the evaluation loss

            # Print the progress for the current epoch, showing the training loss
            print('{} epoch. Train Loss: {}'.format(epoch, eval_loss.item()))

            # Store the evaluation loss for this epoch
            self.Train_loss.append(eval_loss.item())
        
        # After training, plot the training loss
        self.plot_training()
        
    def plot_training(self):
        '''
        Plots the training loss over the course of the training epochs.

        This method visualizes how the loss changes during each epoch of training using a simple line plot. 
        The x-axis represents the epoch number, and the y-axis represents the Mean Squared Error (MSE) loss.

        Returns:
            None
        '''
        
        # Create a figure with 1x1 grid and set the figure size
        fig, ax = plt.subplots(1, 1, figsize=(10, 5))

        # Plot the training loss over the epochs
        ax.plot(range(1, len(self.Train_loss) + 1), torch.tensor(self.Train_loss), label='Train Loss', color='blue')

        # Add a title to the plot
        ax.set_title('Training Loss')

        # Label the x-axis as 'Epoch' and the y-axis as 'Loss'
        ax.set_xlabel('Epoch')
        ax.set_ylabel('Loss')

        # Add a legend to identify the plot (in case there are multiple lines)
        ax.legend()

        # Display the plot
        plt.show()

    def predict(self, input_):
        '''
        Makes a prediction using the trained Black-Scholes neural network model.

        This method takes in an input tensor, processes it using the network, and returns the model's prediction.
        The prediction is made without calculating gradients (i.e., in evaluation mode).

        Args:
            input_ (torch.Tensor): Input tensor containing the feature set for which to predict the options prices.
        
        Returns:
            torch.Tensor: The predicted output (options prices) from the neural network.
        '''
        
        # Disable gradient calculations for prediction (inference)
        with torch.no_grad():
            # Move the input to the appropriate device (CPU or GPU)
            input_ = input_.to(self.device)

            # Make a prediction using the network
            prediction = self.network(input_)
        
        # Return the predicted value(s)
        return prediction
```

```python
# Custom Layer for Subtraction
class SubLayer(nn.Module):
    '''
    Custom PyTorch layer to perform element-wise subtraction between two tensors.
    
    Methods:
        forward(x1, x2): Subtracts tensor x2 from tensor x1.
    '''
    def forward(self, x1, x2):
        return torch.sub(x1, x2)  # Alternatively: x1 - x2

# Custom Layer for Addition
class AddLayer(nn.Module):
    '''
    Custom PyTorch layer to perform element-wise addition between two tensors.
    
    Methods:
        forward(x1, x2): Adds tensor x1 to tensor x2.
    '''
    def forward(self, x1, x2):
        return torch.add(x1, x2)

# Custom Layer for Multiplication
class MulLayer(nn.Module):
    '''
    Custom PyTorch layer to perform element-wise multiplication between two tensors.
    
    Methods:
        forward(x1, x2): Multiplies tensor x1 by tensor x2.
    '''
    def forward(self, x1, x2):
        return torch.mul(x1, x2)

# Custom Layer for Logarithm
class LogLayer(nn.Module):
    '''
    Custom PyTorch layer to compute the natural logarithm of a tensor.
    
    Methods:
        forward(x): Computes the natural log of tensor x.
    '''
    def forward(self, x):
        return torch.log(x)

# Custom Layer for Square Root
class SqrtLayer(nn.Module):
    '''
    Custom PyTorch layer to compute the element-wise square root of a tensor.
    
    Methods:
        forward(x): Computes the square root of tensor x.
    '''
    def forward(self, x):
        return torch.sqrt(x)

# Custom Layer for Squaring
class SqrLayer(nn.Module):
    '''
    Custom PyTorch layer to compute the element-wise square of a tensor.
    
    Methods:
        forward(x): Squares tensor x.
    '''
    def forward(self, x):
        return torch.pow(x, 2)  # Alternatively: x ** 2

# Custom Layer for Division by 2
class DivByTwoLayer(nn.Module):
    '''
    Custom PyTorch layer to divide each element in a tensor by 2.
    
    Methods:
        forward(x): Divides tensor x by 2.
    '''
    def forward(self, x):
        return torch.div(x, 2)  # Alternatively: x / 2

# Custom Layer for Division
class DivLayer(nn.Module):
    '''
    Custom PyTorch layer to perform element-wise division between two tensors.
    
    Methods:
        forward(x1, x2): Divides tensor x1 by tensor x2.
    '''
    def forward(self, x1, x2):
        return torch.div(x1, x2)  # Alternatively: x1 / x2

# Custom Layer for Normal CDF using GELU approximation
class NormalCDFLayer(nn.Module):
    '''
    Custom PyTorch layer to compute the Normal CDF (Cumulative Distribution Function).
    Uses the GELU function as an approximation to the CDF.

    Methods:
        forward(x): Approximates the normal CDF using GELU.
    '''
    def __init__(self):
        super(NormalCDFLayer, self).__init__()
        self.gelu = nn.GELU()  # GELU activation approximates the CDF

    def forward(self, x):
        return torch.div(self.gelu(x), x)  # Approximates normal CDF

# Custom Layer for Exponential
class ExpLayer(nn.Module):
    '''
    Custom PyTorch layer to compute the element-wise exponential of a tensor.
    
    Methods:
        forward(x): Computes the exponential of tensor x (e^x).
    '''
    def forward(self, x):
        return torch.exp(x)


class Black_Scholes_Call_Put_Network(nn.Module):
    '''
    Neural network model for calculating Black-Scholes option pricing for both call and put options.
    The model uses several custom layers to replicate the Black-Scholes mathematical formula.

    Attributes:
        device (torch.device): Device on which computations will be performed (CPU or GPU).
        add_layer (AddLayer): Custom layer for addition operation.
        mul_layer (MulLayer): Custom layer for multiplication operation.
        log_layer (LogLayer): Custom layer for logarithmic operation.
        sqrt_layer (SqrtLayer): Custom layer for square root operation.
        sqr_layer (SqrLayer): Custom layer for squaring operation.
        div2_layer (DivByTwoLayer): Custom layer for division by 2.
        div_layer (DivLayer): Custom layer for division operation.
        sub_layer (SubLayer): Custom layer for subtraction operation.
        normal_cdf_layer (NormalCDFLayer): Custom layer for normal cumulative distribution function approximation.
        exp_layer (ExpLayer): Custom layer for exponential operation.
    '''

    def __init__(self):
        '''
        Initializes the Black-Scholes option pricing network and sets up the layers.
        It determines the device (GPU or CPU) to perform computations.
        '''
        super(Black_Scholes_Call_Put_Network, self).__init__()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        
        # Initialize the custom mathematical operation layers
        self.add_layer = AddLayer()
        self.mul_layer = MulLayer()
        self.log_layer = LogLayer()
        self.sqrt_layer = SqrtLayer()
        self.sqr_layer = SqrLayer()
        self.div2_layer = DivByTwoLayer()
        self.div_layer = DivLayer()
        self.sub_layer = SubLayer()
        self.normal_cdf_layer = NormalCDFLayer()
        self.exp_layer = ExpLayer()

    def forward(self, input_):
        '''
        Performs a forward pass through the network to compute the call and put option prices.
        
        Args:
            input_ (torch.Tensor): A tensor of shape (batch_size, 5), where each row contains:
                                   [S, K, r, T, sigma] representing the following inputs:
                                   - S: Stock price
                                   - K: Strike price
                                   - r: Risk-free interest rate
                                   - T: Time to maturity
                                   - sigma: Volatility

        Returns:
            torch.Tensor: A tensor of shape (batch_size, 2) containing call and put option prices.
        '''
        # Move input to the appropriate device (GPU or CPU)
        input_ = input_.to(self.device)

        # Reshape the input tensor for batch processing
        try:
            input_ = input_.reshape(input_.shape[0], 1, 5)  # For batches
        except: # For prediction
            input_ = input_.view(1, 1, 5)  # For single data point prediction

        # Extract individual components from the input tensor: 
        # S (stock price), K (strike price), r (risk-free rate), T (time to maturity), sigma (volatility)
        S = input_[:, :, 0]
        K = input_[:, :, 1]
        r = input_[:, :, 2]
        T = input_[:, :, 3]
        sigma = input_[:, :, 4]

        # Calculate the intermediate values based on the Black-Scholes formula
        S_K = self.div_layer(S, K)  # S / K
        logS_K = self.log_layer(S_K)  # log(S / K)

        sigma2 = self.sqr_layer(sigma)  # sigma^2
        sigma2_2 = self.div2_layer(sigma2)  # sigma^2 / 2
        
        r_sigma2_2 = self.add_layer(r, sigma2_2)  # r + sigma^2 / 2
        T_r_sigma2_2 = self.mul_layer(r_sigma2_2, T)  # (r + sigma^2 / 2) * T

        logS_K_T_r_sigma2_2 = self.add_layer(logS_K, T_r_sigma2_2)  # log(S / K) + (r + sigma^2 / 2) * T

        T2 = self.sqrt_layer(T)  # sqrt(T)
        sigma_T2 = self.mul_layer(sigma, T2)  # sigma * sqrt(T)

        # Calculate d1 and d2 for the Black-Scholes model
        d1 = self.div_layer(logS_K_T_r_sigma2_2, sigma_T2)  # d1
        d2 = self.sub_layer(d1, sigma_T2)  # d2 = d1 - sigma * sqrt(T)

        # Compute the normal CDF values for d1 and d2 using the GELU approximation
        norm_cdf_d1 = self.normal_cdf_layer(d1)
        norm_cdf_d2 = self.normal_cdf_layer(d2)

        # Compute Call option price using the Black-Scholes formula
        Snorm_cdf_d1 = self.mul_layer(S, norm_cdf_d1)  # S * N(d1)
        Knorm_cdf_d2 = self.mul_layer(K, norm_cdf_d2)  # K * N(d2)
        _rT = self.mul_layer(-r, T)  # -r * T
        exp_rT = self.exp_layer(_rT)  # e^(-rT)
        exp_rTKnorm_cdf_d2 = self.mul_layer(exp_rT, Knorm_cdf_d2)  # e^(-rT) * K * N(d2)

        # If T == 0, return max(S - K, 0) as the Call price, else use the Black-Scholes formula
        Call = torch.where(T == 0, torch.maximum(S - K, torch.tensor(0.0)),
                           self.sub_layer(Snorm_cdf_d1, exp_rTKnorm_cdf_d2))  # Call price

        # Compute Put option price using the Black-Scholes formula
        norm_cdf__d1 = self.normal_cdf_layer(-d1)  # N(-d1)
        norm_cdf__d2 = self.normal_cdf_layer(-d2)  # N(-d2)
        Snorm_cdf__d1 = self.mul_layer(S, norm_cdf__d1)  # S * N(-d1)
        Knorm_cdf__d2 = self.mul_layer(K, norm_cdf__d2)  # K * N(-d2)
        exp_rTKnorm_cdf__d2 = self.mul_layer(exp_rT, Knorm_cdf__d2)  # e^(-rT) * K * N(-d2)

        # If T == 0, return max(K - S, 0) as the Put price, else use the Black-Scholes formula
        Put = torch.where(T == 0, torch.maximum(K - S, torch.tensor(0.0)),
                          self.sub_layer(exp_rTKnorm_cdf__d2, Snorm_cdf__d1))  # Put price

        # Concatenate the Call and Put prices into the final output tensor
        output = torch.cat((Call, Put), dim=1)  # Concatenate Call and Put prices
        output = output.reshape(input_.shape[0], 2)  # Reshape the output to (batch_size, 2)
        
        return output  # Return Call and Put prices
```

```python
# Create an instance of the Black_Scholes_Call_Put class
BS = Black_Scholes_Call_Put()

# Train the model using the generated data
BS.train()
```
